```python
init_url = 'http://bmh.webzdarma.cz/'
```

```python
music_titles = []
```

```python
def retrieve_midi(url, album_title):
    with urlopen(url) as response:
        soup = BeautifulSoup(response, 'html.parser')
        all_links = soup.find_all('a')
        for link in all_links:
            href = link.get('href', '/')
            title = link.getText().strip()
            full_link = init_url + href
            if ".MID" in full_link:
                try:
                    urlretrieve(full_link, path + title + ".MID")
                    music_titles.append(title)
                except:
                    print(url, album_title, title)
```

```python
with urlopen(url) as response:
    soup = BeautifulSoup(response, 'html.parser')
    l = soup.prettify().split('</blockquote>')
    albums = l[2]
    all_links = BeautifulSoup(albums, 'html.parser').find_all("a")
    for link in all_links[:16]:
        href = link.get('href', '/')
        album_title = link.getText().strip()
        final_link = init_url + href
        retrieve_midi(final_link, album_title)
```
